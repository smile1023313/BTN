```python
import time
lenofsize = len(nameline)
lenofsize
```

```python
## find the "type", "mac-address", "id", "RSSI", 
##          "ManufacturerData Key(MDK)", 
##          "ManufacturerData Value(MDV)" 
## And then put them in the DataFrame


namedf=pd.DataFrame(data=np.random.randint(low=0,high=1,size=(lenofsize,6)),
                      columns=['type','mac address','id','RSSI','MDK','MDV'])
```

```python
strat = time.time()


count =0


for num in range(lenofsize):
    print count
#for num in range(10):
##type=0, mac=1, id=2, RSSI=3, MDK=4,MDV=5
    ansp = nameline[num].split(' ',2)
## build unique mac address correspond to row of dataframe
    for find in range(count):
        if namedf['mac address'][find] != ansp[1]:
        ## if mac address is not existence, save in new column
            namedf['type'][num] = ansp[0]
            namedf['mac address'][num] = ansp[1]
            namedf['id'][num] = ansp[2].rstrip('\n')
        
        elif namedf['mac address'][find] == ansp[1]:
        ## if mac address is existence, save in existing column 
            if ansp[2].rstrip('\n').find('RSSI'):
                RSSi = ansp[2].split(' ')
                namedf['RSSI'][find] = RSSi[1]
            elif ansp[2].rstrip('\n').find('Key'):
                key = ansp[2].split(' ')
                namedf['RSSI'][find] = key[2]
            elif ansp[2].rstrip('\n').find('Value'):
                value = ansp[2].split(' ')
                namedf['RSSI'][find] = value[2]
    count+=1

end = time.time()
print end-start
namedf
```
